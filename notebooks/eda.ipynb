```python
data['DateTime'] = pd.to_datetime(data['Time of Occurrence'], dayfirst=True)
data['Hour'] = data['DateTime'].dt.hour
```

```python
def filter_time_window(df, start_hour, end_hour):
    if start_hour < end_hour:
        return df[(df['Hour'] >= start_hour) & (df['Hour'] < end_hour)]
    else:
        # handles windows like 10PM â†’ 2AM
        return df[(df['Hour'] >= start_hour) | (df['Hour'] < end_hour)]

```

```python
def group_by_city(df):
    return (
        df.groupby('City')
          .size()
          .reset_index(name='Count')
    )
```

```python
def add_coordinates(df):
    df['lat'] = df['City'].map(lambda x: city_coords[x][0])
    df['lon'] = df['City'].map(lambda x: city_coords[x][1])
    return df


```

```python
import folium
from folium.plugins import HeatMap

def create_heatmap(df):
    heat_data = df[['lat','lon','Count']].dropna().values.tolist()
    m = folium.Map(location=[23.0,80.0], zoom_start=5)
    HeatMap(heat_data, radius=25, blur=20).add_to(m)
    return m
```

```python
# filter
midnight_crimes = filter_time_window(data, 22, 2)

# group by city
midnight_grouped = group_by_city(midnight_crimes)

# add coordinates
midnight_map_data = add_coordinates(midnight_grouped)

# generate map
midnight_map = create_heatmap(midnight_map_data)
midnight_map

```

```python
##Crimes against women
women_df = data[data['Victim Gender'] == 'F']

women_grouped = group_by_city(women_df)

women_map_data = add_coordinates(women_grouped)

women_heatmap = create_heatmap(women_map_data)
women_heatmap


```

```python
women_night_df = filter_time_window(women_df, 22, 2)

women_night_grouped = group_by_city(women_night_df)

women_night_map = create_heatmap(add_coordinates(women_night_grouped))

women_night_map
```
